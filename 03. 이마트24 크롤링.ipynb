```python
import requests
from bs4 import BeautifulSoup
import pandas as pd




url = "https://www.emart24.co.kr/store"
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

stores = []

for store in soup.find_all('li', class_='searchResult'):
    name = store.find('span', class_='title').text.strip()
    address = store.find('div', class_='place').find('span').text.strip()
    
    if address.startswith('서울특별시'):
        stores.append({'매장명': name, '주소': address})

df = pd.DataFrame(stores)
df.to_csv('seoul_emart24_stores.csv', index=False, encoding='utf-8-sig')

print(f"{len(stores)}개의 서울 이마트24 매장 정보가 seoul_emart24_stores.csv 파일로 저장되었습니다.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://www.emart24.co.kr/store")

# 서울특별시 선택
wait = WebDriverWait(driver, 10)
sido_dropdown = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".sidoWrap .placeHolderWrap")))
sido_dropdown.click()

seoul_option = wait.until(EC.element_to_be_clickable((By.XPATH, "//li[@class='value' and text()='서울특별시']")))
seoul_option.click()

# 검색 버튼 클릭
search_button = driver.find_element(By.CSS_SELECTOR, ".searchBtn")
search_button.click()

# 결과 수집
stores = []
while True:
    time.sleep(2)  # 페이지 로딩 대기
    store_elements = driver.find_elements(By.CSS_SELECTOR, ".searchResult")
    
    for store in store_elements:
        name = store.find_element(By.CSS_SELECTOR, ".title").text
        address = store.find_element(By.CSS_SELECTOR, ".place span").text
        stores.append({"name": name, "address": address})
    
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "div.nextButtons div.next")
        if "opacity: 0.2" in next_button.get_attribute("style"):
            break
        next_button.click()
    except:
        break

driver.quit()

# 결과 출력
for store in stores:
    print(f"매장명: {store['name']}, 주소: {store['address']}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

driver = webdriver.Chrome()
driver.get("https://www.emart24.co.kr/store")

# 서울특별시 선택
wait = WebDriverWait(driver, 10)
sido_dropdown = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".sidoWrap .placeHolderWrap")))
sido_dropdown.click()

seoul_option = wait.until(EC.element_to_be_clickable((By.XPATH, "//li[@class='value' and text()='서울특별시']")))
seoul_option.click()

# 검색 버튼 클릭
search_button = driver.find_element(By.CSS_SELECTOR, ".searchBtn")
search_button.click()

# 결과 수집
stores = []
page_number = 1

while page_number <= 24:
    print(f"페이지 {page_number} 크롤링 중...")
    time.sleep(2)  # 페이지 로딩 대기
    store_elements = driver.find_elements(By.CSS_SELECTOR, ".searchResult")
    
    for store in store_elements:
        name = store.find_element(By.CSS_SELECTOR, ".title").text
        address = store.find_element(By.CSS_SELECTOR, ".place span").text
        stores.append({"name": name, "address": address})
        print(f"매장명: {name}, 주소: {address}")
    
    print(f"페이지 {page_number} 완료\n")
    
    if page_number == 24:
        break
    
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "div.nextButtons div.next")
        if "opacity: 0.2" in next_button.get_attribute("style"):
            break
        next_button.click()
        page_number += 1
    except:
        break

driver.quit()

# CSV 파일로 저장
with open('emart24_stores.csv', 'w', newline='', encoding='utf-8-sig') as file:
    writer = csv.DictWriter(file, fieldnames=['name', 'address'])
    writer.writeheader()
    writer.writerows(stores)

print(f"총 {len(stores)}개의 매장 정보를 수집했습니다.")
print("데이터가 emart24_stores.csv 파일로 저장되었습니다.")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import csv

# Selenium WebDriver 초기화 (Chrome 사용)
driver = webdriver.Chrome()  # ChromeDriver 경로를 지정해야 할 수도 있습니다.

# 이마트24 매장찾기 페이지로 이동
driver.get("https://www.emart24.co.kr/store")

# 명시적 대기를 위한 WebDriverWait 객체 생성 / 조건 만족될때까지 최대 20초 대기 
wait = WebDriverWait(driver, 10)

# 서울특별시 선택
# 시/도 드롭다운 메뉴 클릭 / EC.element_to_be_clickable은 요소가 클릭 가능한 상태까지 기다리게 함 
# By.CSS_SELECTOR 로 CSS 선택자를 사용해 요소를 찾고, 시도 드롭다운 변수에 저장
sido_dropdown = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, ".sidoWrap .placeHolderWrap")))
sido_dropdown.click()

# 서울특별시 옵션 선택 / XPath를 사용해 '서울특별시' 텍스트를 가진 <li> 요소를 찾음
seoul_option = wait.until(EC.element_to_be_clickable((By.XPATH, "//li[@class='value' and text()='서울특별시']")))
seoul_option.click()

# 검색 버튼 클릭
search_button = driver.find_element(By.CSS_SELECTOR, ".searchBtn")
search_button.click()

# 결과 수집을 위한 리스트 초기화
stores = []
page_number = 1

# 페이지별로 데이터 수집
while page_number <= 24:  # 최대 24페이지까지 수집
    print(f"페이지 {page_number} 크롤링 중...")
    time.sleep(2)  # 페이지 로딩 대기

    # 현재 페이지의 매장 정보 수집
    store_elements = driver.find_elements(By.CSS_SELECTOR, ".searchResult")
    
    for store in store_elements:
        name = store.find_element(By.CSS_SELECTOR, ".title").text
        address = store.find_element(By.CSS_SELECTOR, ".place span").text
        stores.append({"name": name, "address": address})
        print(f"매장명: {name}, 주소: {address}")
    
    print(f"페이지 {page_number} 완료\n")
    
    if page_number == 24:
        break
    
    # 다음 페이지로 이동 /버튼이 활성화(투명도 0.2가 아니면)일때 클릭해서 다음페이지로 이동  
    try:
        next_button = driver.find_element(By.CSS_SELECTOR, "div.nextButtons div.next")
        if "opacity: 0.2" in next_button.get_attribute("style"):
            break  # 다음 페이지 버튼이 비활성화되면 종료
        next_button.click()
        page_number += 1
    except:
        break  # 다음 페이지 버튼을 찾지 못하면 종료

# 브라우저 종료
driver.quit()

# 수집한 데이터를 CSV 파일로 저장  /with open(..)은 파일열고작업이끝나면 자동닫는 파이썬관리자
# 'w'는 파일열기모드(쓰기모드, 기존파일이 있으면 덮어씀) /as file은 열린 파일 객체를 file변수에 할당 
# newline은 줄바꿈을 자동으로 처리하는 것 / fieldnames=는 이름, 주소 두개의 열을 만듬 

with open('emart24_stores.csv', 'w', newline='', encoding='utf-8-sig') as file:
    writer = csv.DictWriter(file, fieldnames=['name', 'address'])
    writer.writeheader()  # 열이름 : 파일 첫번째 행에 열이름(헤더)를 쓴다
    writer.writerows(stores)  # stores 리스트의 모든 딕셔너리(매장정보)를 csv파일에 행으로 쓴다.  

print(f"총 {len(stores)}개의 매장 정보를 수집했습니다.")
print("데이터가 emart24_stores.csv 파일로 저장되었습니다.")
```
